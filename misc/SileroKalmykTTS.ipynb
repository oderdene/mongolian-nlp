# Kalmyk TTS with Silero

This notebook uses Silero from the project [snakers4/silero-models](https://github.com/snakers4/silero-models/) to synthesize a Kalmyk text.

## Install snakers4/silero-models

```
#@title

!pip install -q torchaudio omegaconf

import torch
from pprint import pprint
from omegaconf import OmegaConf
from IPython.display import Audio, display

torch.hub.download_url_to_file('https://raw.githubusercontent.com/snakers4/silero-models/master/models.yml',
                               'latest_silero_models.yml',
                               progress=False)
models = OmegaConf.load('latest_silero_models.yml')

language = 'xal'
speaker = 'erdni_v2'
sample_rate = 16000
device = torch.device('cpu')
model, example_text = torch.hub.load(repo_or_dir='snakers4/silero-models',
                                     model='silero_tts',
                                     language=language,
                                     speaker=speaker)
model.to(device)  # gpu or cpu
```

## Synthesize

```
#@markdown { run: "auto" }

text = "\u042D\u043D\u0442\u043D \u0434\u0435\u0433\u0434 \u0441\u04D9\u04D9\u0445\u043D \u0448\u04AF\u043B\u0433 \u0431\u043E\u043B\u0497, \u0433\u0438\u0497 \u0431\u0438 \u04AF\u043D\u043D \u0441\u0435\u0434\u043A\u043B\u04D9\u0441\u043D \u043A\u0435\u043B\u04AF\u0432." #@param {type:"string"}
audio = model.apply_tts(texts=[text], sample_rate=sample_rate)

print(text)
display(Audio(audio[0], rate=sample_rate))
```
